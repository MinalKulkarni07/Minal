```python
# Q1.
#-- User input section
#Enter Product List Size : 4
#Enter Product 1 -
#Samosa
#Enter Product 2 -
#Kachori
#Enter Product 3 -
#Fafda
#Enter Product 4 -
#Jalebi

#Enter Samosa Price :
#300/-
#Enter Kachori Price :
#100/-
#Enter Fafda Price :
#100/-
#Enter Jalebi Price :
#200/-

#--- output

#samosa = 300
#kachori = 100
#fafda = 100
#jalebi = 200
```

```python
# Q1.- 
size = int(input("Enter product list size: "))

product = []  
price = []

for i in range(1, size + 1):
    prod = input("Enter product " + str(i) + ": ")
    product.append(prod)  
    rate = input("Enter " + prod + " price : ")
    price.append(str(int(rate)))


print("\n--- Product List ---")
for i in range(size):
    print(product[i]+ " = " +price[i])
```

```python
# Q2.
# -- User input section
#Enter Product List Size : 4
#Enter Product 1 -
#Samosa
#Enter Product 2 -
#Kachori
#Enter Product 3 -
#Fafda
#Enter Product 4 -
#Jalebi

#Enter Samosa Price :
#300/-
#Enter Kachori Price :
#100/-
#Enter Fafda Price :
#100/-
#Enter Jalebi Price :
#200/-

#Do You Want to add GST [Y/N]
#[note : if user select `N` Print bill without GST]
#[Note User select yes `Y`]
#Enter GST % = 18
#-------Out put -------

#1. Samosa = 300/-
#2. Kachori = 100/-
#3. Fafda = 100/-
#4. Jalebi = 200/-
#-------------------
#Total = 700
#GST = 18%
#-------------------
#Final Total = 826/-
#-------------------
```

```python
# Q2. With gst
size = int(input("Enter product list size: "))

product = []  
price = []

for i in range(1, size + 1):
    prod = input("Enter product " + str(i) + ": ")
    product.append(prod)  
    rate = input("Enter " + prod + " price : ")
    price.append(str(int(rate)))

def total(price):
    return sum(int(p)for p in price)
    
def gst(Total, gst_percentage):
    return Total*(gst_percentage/100)
    
def final_total(Total, GST):
    return (Total + GST)
    
Total = total(price)    

Add_GST = input("Do you want to add GST [Y/N]").strip().upper()
GST = 0           
if (Add_GST == "Y"):
    gst_percentage = float(input("Enter GST percentage :"))
    GST = round(gst(Total, gst_percentage), 2)
    print("GST =" + str(GST))
elif (Add_GST == "N"):
    print("_")

Final_Total = final_total(Total, GST)

print("\n--- Total Bill ---")
for i in range(size):
    print(str(i+1)+". "+ product[i]+ " = " +str(price[i]))
print("\n------------------")
print("Total =" + str(Total))
print("GST =" + str(gst_percentage))
print("\n------------------")
print("Final Total =" + str(Final_Total))
```
